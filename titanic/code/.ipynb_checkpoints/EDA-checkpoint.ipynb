#  2. 탐색적 데이터분석(EDA)과 전처리 그리고 파생변수 생성

**<목적>**

* 분석할 때 데이터를 잘 이해할 수록 좋은 분석결과가 나옵니다. 따라서 EDA를 통해 데이터를 완벽하게 이해하는 것이 가장 이상적입니다. 하지만 시간과 도메인이라는 한계가 존재하고, 경험상 데이터를 한번에 이해하려 하기보다. 한번의 모델링 후 이해하는 것이 더 효율적이라는 사실을 깨달았습니다.

* 탐색적 이라는 단어를 조금 생각해 봐야합니다. 우리가 탐색한다는 이미지를 생각한다면 목표를 향해 한 길만 올곧게 가기보다. 여기저기 살펴보는 이미지를 떠올릴것입니다. 탐색적 데이터분석도 마찬가지 입니다. 데이터의 분포를 다방면으로 살펴보고 나눠보고 합쳐보면서 데이터에 대한 인사이트를 얻게됩니다.

### (3) 칼럼 특성과 분포 , 타겟 변수와의 관계 파악

**<구성>**

1. 범주형/ 카테고리 변수 구분 후 각 변수들간 분포확인
    - 제 경우는 (1) 작업효율 (2) 동일한 변수성질끼리 비교하기 편하기 때문에, 라는 두 가지 이유로 구분 후 진행합니다.
    - 순차적으로 분석해도 됩니다. 분석가가 편한 방법으로 진행하셔도 무방하다고 생각합니다.
    
2. 종속변수와 타겟변수간 관계
3. 종속변수들간의 관계

### 범주형 변수

1. Sex
2. Embarked
3. Pclass
4. SibSp
5. Parch
6. Cabin

```python
fig, ax = plt.subplots(2,3, figsize=(20,15))
pd.crosstab(train.Sex, train.Survived, normalize='index').plot(kind='bar', 
                                                               ax=ax[0,0], 
                                                               stacked=True,
                                                               alpha =0.7)

pd.crosstab(train.Embarked, train.Survived, normalize='index').plot(kind='bar', 
                                                                    ax=ax[0,1], 
                                                                    stacked=True,
                                                                    alpha =0.7)

pd.crosstab(train.Pclass, train.Survived, normalize='index').plot(kind='bar', 
                                                                  ax=ax[0,2], 
                                                                  stacked=True,
                                                                  alpha =0.7)

pd.crosstab(train.SibSp, train.Survived, normalize='index').plot(kind='bar', 
                                                                 ax=ax[1,0], 
                                                                 stacked=True,
                                                                 alpha =0.7)

pd.crosstab(train.Parch, train.Survived, normalize='index').plot(kind='bar', 
                                                                 ax=ax[1,1], 
                                                                 stacked=True,
                                                                 alpha =0.7)

plt.tight_layout()
plt.show()
```

### Cabin 변수

```python
# 결측치 제외 후 살펴본다.
fig, ax = plt.subplots(1,1,figsize=(30,10))
pd.crosstab(train.Cabin.dropna(), train.Survived, normalize='index').plot(kind='bar',
                                                                          ax=ax,
                                                                          alpha =0.7)
plt.show()
```

```python
fig, ax = plt.subplots(1,1,figsize=(30,10))
pd.crosstab(train.Cabin, train.Survived).plot(kind='bar',ax=ax,alpha =0.7)
plt.show()
```

### 연속형 변수

```python
def vis_diff(df, col):
    yes = df.loc[df.Survived==1].dropna()
    no = df.loc[df.Survived==0].dropna()
    fig, (ax1, ax2) = plt.subplots(
                               1,2, 
                               figsize=(15, 5),
                               facecolor='w', 
                               edgecolor='k'
                              )

    sns.distplot(yes[col], ax=ax1, color='blue',  label='Survived YES')
    sns.distplot(no[col], ax=ax1, color='red', label='Survived NO')
    ax1.legend()
    ax1.set(title='PDF of Survived')

    N = yes[col].shape[0]
    X2 = np.sort(yes[col])
    F2 = np.array(range(N))/float(N)

    yes_plot = ax2.plot(X2, F2, alpha=0.8, color='blue', label='Survived YES')

    ax2.set(title="CDF of Survived")

    N = no[col].shape[0]
    X2 = np.sort(no[col])
    F2 = np.array(range(N))/float(N)

    no_plot = ax2.plot(X2, F2, alpha=0.8, color='red', label='Survived NO')
    ax2.legend()

    fig.tight_layout()
    plt.show()
```

### Age 변수

```python
vis_diff(train, 'Age')
```

```python
# 나이 소수점을 버리고 Int형으로 변환
train.Age = round(train.Age).astype(np.int)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
pd.crosstab(train.Age, train.Survived, normalize='index').plot(kind='bar', 
                                                               stacked=True,
                                                               ax=ax,
                                                               alpha =0.3)
plt.show()
```

### Fare 변수

```python
vis_diff(train, 'Fare')
```

```python
train.Fare = round(train.Fare)
```

```python
fig, ax = plt.subplots(1,1,figsize=(20,10))
pd.crosstab(train.Fare, train.Survived, normalize='index').plot(kind='bar', 
                                                               stacked=True,
                                                               ax=ax,
                                                               alpha =0.3)
plt.show()
```
